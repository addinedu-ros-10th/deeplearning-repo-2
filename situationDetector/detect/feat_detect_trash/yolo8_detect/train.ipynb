```python
import yaml
from ultralytics import YOLO
import os

# --- ì¤‘ìš” ---
# ì´ ìŠ¤í¬ë¦½íŠ¸ë¥¼ ì‹¤í–‰í•˜ê¸° ì „ì—, 'download_dataset.py'ë¥¼ ë¨¼ì € ì‹¤í–‰í•˜ì—¬
# ë°ì´í„°ì…‹ í´ë”ê°€ ìƒì„±ë˜ì—ˆëŠ”ì§€ í™•ì¸í•´ì•¼ í•©ë‹ˆë‹¤.

# 2. ë°ì´í„°ì…‹ ì„¤ì • íŒŒì¼(YAML) ê²½ë¡œ í™•ì¸
# Roboflowì—ì„œ ë‹¤ìš´ë¡œë“œí•˜ë©´ ìë™ìœ¼ë¡œ ìƒì„±ë˜ëŠ” ì„¤ì • íŒŒì¼ì…ë‹ˆë‹¤.
DATASET_DIR = '/home/momo/Desktop/dataset/thesis-kztn8_littering-whlsk(8)/Littering--8' #<-- 'littering-whlsk-1'ì—ì„œ ìˆ˜ì •ë¨
data_yaml_path = os.path.join(DATASET_DIR, 'data.yaml')

# YAML íŒŒì¼ ë‚´ìš©ì„ í™•ì¸í•˜ì—¬ ê²½ë¡œì™€ í´ë˜ìŠ¤ê°€ ì˜¬ë°”ë¥¸ì§€ ì‚´í´ë´…ë‹ˆë‹¤.
try:
    with open(data_yaml_path, 'r') as f:
        data_yaml = yaml.safe_load(f)
        print("--- data.yaml íŒŒì¼ ë‚´ìš© ---")
        print(data_yaml)
        print("---------------------------\n")
        
        # Check if paths exist before training
        for split in ['train', 'val', 'test']:
            if data_yaml.get(split):
                abs_path = os.path.abspath(os.path.join(DATASET_DIR, data_yaml[split]))
                data_yaml[split] = abs_path
                print(f"Checking {split} path: {abs_path}")
                if not os.path.exists(abs_path):
                    print(f"Error: {split} path does not exist: {abs_path}")
                    exit(1)
        
        with open(data_yaml_path, 'w') as f:
            yaml.dump(data_yaml, f)
            print("Updated data.yaml with absolute paths.")

except FileNotFoundError:
    print(f"ì˜¤ë¥˜: '{data_yaml_path}' íŒŒì¼ì„ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤.")
    print("ë°ì´í„°ì…‹ í´ë” ì´ë¦„ì´ 'Littering--8'ì´ ë§ëŠ”ì§€, 'download_dataset.py'ë¥¼ ë¨¼ì € ì‹¤í–‰í–ˆëŠ”ì§€ í™•ì¸í•´ì£¼ì„¸ìš”.")
    exit(1)

# 3. YOLOv8 ëª¨ë¸ ë¶ˆëŸ¬ì˜¤ê¸°
# ê°€ì¥ ì‘ê³  ë¹ ë¥¸ yolov8n (nano) ëª¨ë¸ë¡œ ì‹œì‘í•©ë‹ˆë‹¤.
print("YOLOv8n ëª¨ë¸ì„ ë¶ˆëŸ¬ì˜µë‹ˆë‹¤...")
model = YOLO('yolov8n.pt')


# 4. ëª¨ë¸ í›ˆë ¨ ì‹œì‘
# epochs: ì „ì²´ ë°ì´í„°ì…‹ì„ ëª‡ ë²ˆ ë°˜ë³µ í•™ìŠµí• ì§€ ê²°ì • (100ë²ˆìœ¼ë¡œ ì‹œì‘)
# imgsz: í›ˆë ¨ì— ì‚¬ìš©í•  ì´ë¯¸ì§€ í¬ê¸° (640x640)
# patience: ì§€ì •ëœ epoch ë™ì•ˆ ì„±ëŠ¥ í–¥ìƒì´ ì—†ìœ¼ë©´ í›ˆë ¨ì„ ì¡°ê¸° ì¢…ë£Œ (ê³¼ì í•© ë°©ì§€)
print("\nëª¨ë¸ í›ˆë ¨ì„ ì‹œì‘í•©ë‹ˆë‹¤...")
results = model.train(
    data=data_yaml_path,
    epochs=100,
    imgsz=640,
    patience=20,
    name='yolov8n_littering_detector' # í›ˆë ¨ ê²°ê³¼ê°€ ì €ì¥ë  í´ë” ì´ë¦„
)

print("\ní›ˆë ¨ì´ ì™„ë£Œë˜ì—ˆìŠµë‹ˆë‹¤!")
print("ê²°ê³¼ëŠ” 'runs/detect/yolov8n_littering_detector' í´ë”ì— ì €ì¥ë˜ì—ˆìŠµë‹ˆë‹¤.")


# 5. (ì„ íƒ) í›ˆë ¨ëœ ëª¨ë¸ë¡œ í…ŒìŠ¤íŠ¸ ì´ë¯¸ì§€ ì˜ˆì¸¡
# ê°€ì¥ ì„±ëŠ¥ì´ ì¢‹ì•˜ë˜ ê°€ì¤‘ì¹˜(best.pt)ë¥¼ ì‚¬ìš©í•˜ì—¬ ì˜ˆì¸¡ì„ ìˆ˜í–‰í•©ë‹ˆë‹¤.
print("\ní›ˆë ¨ëœ ëª¨ë¸ë¡œ í…ŒìŠ¤íŠ¸ ì˜ˆì¸¡ì„ ìˆ˜í–‰í•©ë‹ˆë‹¤...")
best_model_path = os.path.join('runs', 'detect', 'yolov8n_littering_detector', 'weights', 'best.pt')

# í›ˆë ¨ëœ ëª¨ë¸ì´ ì¡´ì¬í•˜ëŠ”ì§€ í™•ì¸
if os.path.exists(best_model_path):
    best_model = YOLO(best_model_path)
    
    test_images_path = data_yaml.get('test')

    if test_images_path and os.path.exists(test_images_path):
        # save=True ì˜µì…˜ìœ¼ë¡œ ì˜ˆì¸¡ ê²°ê³¼(ì‚¬ê°í˜•ì´ ê·¸ë ¤ì§„ ì´ë¯¸ì§€)ë¥¼ ì €ì¥í•©ë‹ˆë‹¤.
        best_model.predict(
            source=test_images_path,
            save=True,
            name='yolov8n_littering_prediction'
        )

        print("\nì˜ˆì¸¡ì´ ì™„ë£Œë˜ì—ˆìŠµë‹ˆë‹¤!")
        print("ì˜ˆì¸¡ ê²°ê³¼ ì´ë¯¸ì§€ëŠ” 'runs/detect/yolov8n_littering_prediction' í´ë”ì— ì €ì¥ë˜ì—ˆìŠµë‹ˆë‹¤.")
    else:
        print("í…ŒìŠ¤íŠ¸ ì´ë¯¸ì§€ ê²½ë¡œë¥¼ ì°¾ì„ ìˆ˜ ì—†ì–´ ì˜ˆì¸¡ì„ ê±´ë„ˆëœë‹ˆë‹¤.")
else:
    print(f"í›ˆë ¨ëœ ëª¨ë¸ ê°€ì¤‘ì¹˜({best_model_path})ë¥¼ ì°¾ì„ ìˆ˜ ì—†ì–´ ì˜ˆì¸¡ì„ ê±´ë„ˆëœë‹ˆë‹¤.")
```
